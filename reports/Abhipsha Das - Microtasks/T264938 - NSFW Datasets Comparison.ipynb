## Datasets Survey for T264045: Not Safe for Work (NSFW) media Classifier for Wikimedia Commons

NSFW (or Not Safe for Work) is an internet slang that arose out of utility to communicate and indicate a certain type of visual data that may be sexual(explicit or suggestive), violent or harmful content to view in non-casual, professional environments. As such, [Wikipedia](https://www.wikiwand.com/en/Not_safe_for_work) defines it as:

> Not safe for work (NSFW) is Internet slang or shorthand used to mark links to content, videos or websites pages the viewer may not wish to be seen looking at in a public, formal or controlled environment. The marked content may contain nudity, intense sexuality, political incorrectness, profanity, slurs, violence or other potentially disturbing subject matter. Environments that may be problematic include workplaces, schools, and family settings. NSFW has particular relevance for people trying to make personal use of the Internet at workplaces or schools which have policies prohibiting access to sexual and graphic subject matter. 

> Conversely, safe for work (SFW) is used for links that don't contain such material, but where the title might otherwise lead people to think that content is NSFW. 

Hence, it has become a significant endeavour by data scientists, web-based companies and aware individuals alike to work towards improving the identification of NSFW media and content in order to ensure greater safety in while browsing the internet or using web services in public and professional environments. Towards this, a lot of research has been conducted that necessarily requires curation and publication of the datasets that were used.

In this report, I will comparing some of the datasets that are commonly used to train NSFW classifiers, and some lesser known ones as well, all of which are freely accessible or after obtaining permission from the dataset curators. 


### 1. [NPDI Pornography Database](https://sites.google.com/site/pornographydatabase/)

The Pornography database contains nearly 80 hours of 400 pornographic and 400 non-pornographic videos. The database consists of several genres of pornography and depicts actors of many ethnicities, including multi-ethnic ones. For the non-pornographic class, we have browsed general-public purpose video network and selected two samples: 200 videos chosen at random (which we called "easy") and 200 videos selected from textual search queries like "beach", "wrestling", "swimming", which we knew would be particularly challenging for the detector (called "difficult").

### 2. [Pornography-2k dataset](https://recodbr.wordpress.com/code-n-data/#porno)

The Pornography-2k dataset is an extended version of the NPDI dataset. The dataset comprises nearly 140 hours of 1,000 pornographic and 1,000 non-pornographic videos, which varies from six seconds to 33 minutes. Pornography-2k dataset is very assorted, including both professional and amateur content. Moreover, it depicts several genres of pornography, from cartoon to live action, with diverse behavior and ethnicity.

#### Pros:

This appears to be a hugely vast database and will be useful for training NSFW classifiers, on the condition that ample SFW images are integrated and a combined database of both categories is curated. 

#### Cons: 

This dataset is not open-sourced and freely accessible. Researchers interested in getting this dataset have to contact the curators to request access to this dataset.

### 4. [NSFW Data Scraper](https://github.com/alex000kim/nsfw_data_scraper)

This is a open-sourced project hosted on GitHub that utilizes data scrapers to scrape NSFW media. As stated by the author, it is a set of scripts that allows for an automatic collection of tens of thousands of images for the following (loosely defined) categories to be later used for training an image classifier:

    porn - pornography images
    hentai - hentai images, but also includes pornographic drawings
    sexy - sexually explicit images, but not pornography. Think nude photos, playboy, bikini, etc.
    neutral - safe for work neutral images of everyday things and people
    drawings - safe for work drawings (including anime)
    
The individual scrapers can all be run by pulling and running the docker image of this project. After some hours all the data gets downloaded to our local system.

#### Pros: 

It is open-source and easily accessible, requires only a few commands to pull, run and build from docker to have all the data downloaded locally. More scraping scripts can be added to the folders to collaborate and create more extensive datasets.

#### Cons: 

It is a noisy dataset that needs to go through various steps of preprocesing and cleaning before it can be used.

### 7. [GantMan's Dataset](https://github.com/GantMan/nsfw_model)

This is an extensive dataset curated by GantMan and open-sourced on GitHub along with the implementation of their NSFW classifier. The project repository contains website scrapers that automatically download images and store them in the `images` folder when cloned and run locally. The dataset has been dvided into these categories:

    drawings - safe for work drawings (including anime)
    hentai - hentai and pornographic drawings
    neutral - safe for work neutral images
    porn - pornographic images, sexual acts
    sexy - sexually explicit images, not pornography

amounting to 60+ GB of data.

#### Pros:

It is open-sourced and freely accessible. The dataset on downloading is made made by combining SFW images with NSFW content. 

#### Cons: 

The dataset comes with the NSFW classfier implementation. The whole project needs to be cloned and run in order let the scraping scripts download the images and store them locally. The dataset needs to be cleaned, preprocessed and augmented before it can be used.

### Additional Tools

To supplement our findings and create an even more comprehensive data resource, it is useful that we ourselves scrape websites for visual content as well as preprocess and augment them. Towards this end, some popular scraping tools should be used to enhance our database. I have listed a few that can be very useful while curating our own database.


#### 1. [RipMe](https://github.com/RipMeApp/ripme)

RipMe is an open-sourced album ripper for various websites. It is a cross-platform GUI tool that requires Java 8, and can be run on Windows, Linux and MacOS. It quickly downloads images from any online album or image hosting platform after specifying the URL of that website in the application.

List of supported sites:

> imgur, twitter, tumblr, instagram, flickr, photobucket, reddit, gonewild, motherless, imagefap, imagearn, seenive, vinebox, 8muses, deviantart, xhamster

However, RipMe does not support any feature to restrict the amount of data to be downloaded.

#### 2. [gallery-dl](https://github.com/mikf/gallery-dl)

It is also an open-sourced command line tool to download image-galleries and collections from several image hosting sites. It is a cross-platform tool with many configuration options and powerful filenaming capabilities. It requires Python 3.4+ to run. 

List of supported sites can be seen [here](https://github.com/mikf/gallery-dl/blob/master/docs/supportedsites.rst).

#### 3. [Augmentor](https://github.com/mdbloice/Augmentor)

It is an open-source image augmentation library in Python for machine learning. It implements the most real-world relevant augmentation techniques and employs a stochastic approach using building blocks that allow for operations to be pieced together in a pipeline.

It is a very useful library to have to augment datasets, and is used by writing Python scripts and importing the Augmentor library. It uses multi-threading to speed up the augmentation process. 


Other resources and methods like [Gemini](https://macpaw.com/geminihttps://www.snorkel.org) and [image hashing](https://www.pyimagesearch.com/2020/04/20/detect-and-remove-duplicate-images-from-a-dataset-for-deep-learning/) to remove duplicate images can be used to preprocess and prepare the dataset.
